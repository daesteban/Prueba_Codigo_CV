#### 8. Crear columnas con las siguiente información: el % de mujeres que no escriben ni leen, % de varones que no escriben ni leen y % de nativos respecto al total de la población.

```r
base_junin['%nesc_wom']=base_junin$nesc_mujer/base_junin$nesc_total #% de mujeres que no escriben
base_junin['%nlee_wom']=base_junin$nlee_mujer/base_junin$ntotal_lee #% de mujeres que no escriben
base_junin['%nesc_men']=base_junin$nesc_hombre/base_junin$nesc_total #% de hombres que no escriben
base_junin['%nlee_men']=base_junin$nlee_hombre/base_junin$ntotal_lee #% de hombres que no escriben
base_junin['%nativo']=base_junin$natives/(base_junin$whites+base_junin$natives+ base_junin$mestizos+base_junin$blacks) #% de nativos respecto al total de la población
base_junin['pob_tot']=(base_junin$whites+base_junin$natives+base_junin$mestizos+base_junin$blacks) #total de la población
base_junin
```

#### 9. Crear una base de datos con la siguiente información:
- Quedarse con la información de los distritos de Ciudad del Cerro, Jauja, Acolla, San Gerónimo, Tarma, Oroya y Concepción
- Ciudades con una cantidad de niños (4-6 años) mayor a 50 y de (6-14 años) > 70
- Ordenar las variables según el total de la población y cantidad de población nativa
- Solo quedarse con las variables trabajadas en 8, Distrito, y comunidad.
- Guardar la mitad de la información en bases diferentes (llamar a cada uno Base1 y Base 2)
- Guardar la Base1 en formato csv y excel en la carpeta data. (Use el siguiente nombre Base1_WG(numero de grupo))

```r
districts <- base_junin[base_junin$District %in% c("CIUDAD DEL CERRO","JAUJA","ACOLLA","SAN GERÓNIMO","TARMA","OROYA","CONCEPCIÓN"),]
districts
```

```r
cities <- districts %>% filter((districts['4_6_years_men'] > 50) & (districts['6_14_years_men'] > 70))
cities
```

```r
base = head(cities[order((cities$pob_tot),(cities$natives)),])
base
```

```r
new_base = head(base[,c('District','comunidad','%nativo','%nlee_men','%nesc_men', '%nlee_wom')])

new_base
```

#### 10. Crear una función que me permita extraer columnas un rango de columnas y de filas. Crear una función que me permite extraer una columna según el nombre de la variable y dos filas.

```r
funcion1 <- function(data_base, posición_fila1, posición_fila2, posición_columna1, posición_columna2) {
  resultado <- data_base[posición_fila1:posición_fila2,posición_columna1:posición_columna2]
  return(resultado)
}
```
