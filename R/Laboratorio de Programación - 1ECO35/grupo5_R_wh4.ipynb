## Filtro de base de datos

1. Quedarse con los hogares Urbanos, asimismo considerar los hogares que completaron la encuesta.

```r
enapres2020_1$AREA %>% attr('labels')
```

```r
#Solo hogares urbanos
df <- enapres2020_1 %>% filter(enapres2020_1["AREA"]==1)
#Solo hogares que completaron la encuesta
df <- df %>% filter(df["RESFIN"]==1)
head(df)
```

2. Verificar si existe duplicados de hogares

```r
df %>% group_by(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR) %>% filter(n()>1) 
```

3. Si existiese, filtrar la base de datos para no considerar los duplicados.

```r
df_ndup <- df %>% distinct(CCDD ,CCPP , CCDI ,CONGLOMERADO , NSELV, VIVIENDA, HOGAR, .keep_all = TRUE)
```

```r
df_ndup
```

## Crear Dummies

1. Crear una Dummy si la zona donde se ubica el hogar instaló cámaras de seguridad. (variable P122A_1). Tenga cuidado con los missing values. Tenga cuidado con los missing values si los hubiera.

```r
unique(enapres2020_1$P122A_1)
```

```r
df_ndup <- df_ndup %>%  mutate(Dummy_camara_seguridad = ifelse(P122A_1 == 1 ,  1 , ifelse(!is.na(P122A_1),0, NA) ) )
```

2. Crear una Dummy si el hogar cuenta con internet (variable P184A_14). Tenga cuidado con los missing values si los hubiera

```r
unique(enapres2020_1$P184A_14)
```

```r
df_ndup <- df_ndup %>%  mutate(Dummy_internet = ifelse(P184A_14 == 1 ,  1 , ifelse(!is.na(P184A_14),0, NA) ) )
```

3. Crear una Dummy si el hogar cuenta con titulo de propiedad (variable P107A)

```r
unique(enapres2020_1$P107A)
```

```r
df_ndup <- df_ndup %>%  mutate(Dummy_titulo_propiedad = ifelse(P107A == 1 ,  1 , ifelse(!is.na(P107A),0, NA))) 
```

```r
df_ndup
```

## Group by

1. Obtener el total de hogares por distrito, el total de hogares que cuentan con internet por distrito, finalmente el porcentaje de hogares que cuentan con internet.

```r
# Mostrar total de hogares por distrito (hogares_total) y total de hogares con internet (hogares_internet_total)
base_a <- df_ndup %>% group_by(CCDD, CCPP, CCDI) %>% 
                    summarize(hogares_total = n(), hogares_internet_total=sum(Dummy_internet))
base_a
```

```r
# Agregar porcentaje de hogares con internet
base_a$hogares_internet_porcentaje <- (base_a$hogares_internet_total/base_a$hogares_total)*100
base_a
```

2. Obtener una base de datos a nivel departamentos con la siguiente información: total de hogares, total de hogares cuyo vecindario instaló cámaras de seguridad y el porcentaje hogares cuyos vecindarios cuentan con cámaras de seguridad. Base_1

```r
Base_1 <- df_ndup %>% group_by(CCDD) %>%
            summarize(departamento_hogares=n(), hogares_camara_total=sum(!is.na(Dummy_camara_seguridad)))
Base_1
```

```r
Base_1$hogares_camara_porcentaje <- (Base_1$hogares_camara_total/Base_1$departamento_hogares)*100
Base_1
```

3. Crear una base de datos a nivel departamentos con la siguiente información: total de hogares que cuentan con títulos de propiedad y porcentaje de hogares que cuentan con títulos de propiedad. Base_2

```r
Base_2  <- df_ndup %>% group_by(CCDD) %>% 
                    summarize(departamento_hogares = n(), hogares_titulo_total=sum(!is.na(Dummy_titulo_propiedad)))
Base_2
```

```r
Base_2$hogares_titulo_porcentaje <- (Base_2$hogares_titulo_total/Base_2$departamento_hogares)*100
Base_2
```

## Merge

1. Hacer un merge entre las bases de datos Base_1 y Base_2

```r
Clean_data <- merge(Base_1, Base_2, by = c("CCDD","departamento_hogares"), all.x = TRUE)
Clean_data
```

2. Finalmente guardar la base de datos en el formato SPSS. Llamar a la base de datos (Clean_data)

```r
write_sav(Clean_data, "Clean_Data-Grupo_5-R.sav")
```
