```python
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import numpy as np
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',options=chrome_options)
```

**Codigo para conseguir los nombres de los proyectos y los links**

```python
wd.get("https://www.investinperu.pe/es/app/portafolio-de-proyectos-de-proinversion")
projects = wd.find_elements(By.XPATH,"//div[@class='card-body']//a")
projects_description = wd.find_elements(By.XPATH,"//div[@class='portafolio-card']//div[@class='col-xs-12 col-lg-6']")
name = []
links = []
description = []

for lnk in projects:
  name.append(lnk.get_attribute("text"))
  links.append(lnk.get_attribute("href"))

for idx in projects_description:
  description.append(idx.text)

```

Limpieza y manejo de los datos

```python
Monto_Inversion_sinIGV = description[0:len(description):4]
Cartera = description[1:len(description):4]
Modalidad = description[2:len(description):4]
Fase = description[3:len(description):4]

Monto_Inversion_sinIGV = [item.split(': ', 1)[1] for item in Monto_Inversion_sinIGV]
Monto_Inversion_sinIGV = [item.split('$ ', 1)[1] for item in Monto_Inversion_sinIGV]
Monto_Inversion_sinIGV = [item.split(' M', 1)[0] for item in Monto_Inversion_sinIGV]
Monto_Inversion_sinIGV = [s.replace(',','') for s in Monto_Inversion_sinIGV]
Monto_Inversion_sinIGV = [float(x) for x in Monto_Inversion_sinIGV]
Cartera = [item.split(': ', 1)[1] for item in Cartera]
Modalidad = [item.split(': ', 1)[1] for item in Modalidad]
Fase = [item.split(': ', 1)[1] for item in Fase]
```

```python
data = pd.DataFrame({'Proyecto':name,'Monto de Inversion sin IGV (Millones de US$)':Monto_Inversion_sinIGV,'Cartera':Cartera,'Modalidad':Modalidad,'Fase':Fase,'Link':links})
```
