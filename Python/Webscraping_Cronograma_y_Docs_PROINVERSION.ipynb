# **Descripción**

Este código obtiene información sobre los cronogramas y los documentos de cada proyecto de inversión en la [cartera de Proinversión](https://www.investinperu.pe/es/app/portafolio-de-proyectos-de-proinversion).

El próposito del output es elaborar una alarma sobre los nuevos documentos subidos de cada proyecto para estar informado sobre las nuevas actualización de cada proyecto de inversión de APP de PROINVERSIÓN.

**Inputs necesarios:**
* Un excel con los nombres del proyecto y su respectivo link a cada uno.

```python
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import numpy as np
import re
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',options=chrome_options)
```

Importamos datos del excel sobre los proyectos APPs

```python
data = pd.read_excel(r'/content/Cartera de Proyectos Proinversion.xlsx')
data
```

```python
output = []
proyecto = data['Proyecto'].tolist()
links = data['Link'].tolist()
#link = links[3] # proyecto solo ficha y documentos
```

# Datos sobre cronograma y documentos de cada proyecto.

Tiempo estimado: 5 minutos.

```python
for idx in range(0,len(proyecto)):
  link = links[idx]
  proyect = proyecto[idx]
  output.append(proyect)
  time.sleep(5)
  #######
  wd.get(link) #definimos url
  ######
  asd = []
  tab = wd.find_elements(By.XPATH,"//div[@class='portafolio-nav-wrapper']//a[@role='tab']")
  for lnk in tab:
    asd.append(lnk.get_attribute("text")) #nombre de tabs
  #######
  if any("Cronograma" in item for item in asd): #si hay cronograma en la pagina web
    row = wd.find_elements(By.XPATH,"//table[@class='table table-default']/tbody//td") #elementos filas
    col = wd.find_elements(By.XPATH,"//table[@class='table table-default']/thead/tr/th") #elementos columnas
    header = []
    for lnk in col:
      header.append(lnk.get_attribute("innerHTML"))
    asd2 = []
    for lnk in row:
      asd2.append(lnk.get_attribute("innerHTML"))
    first_col = asd2[0:len(asd2):2]
    second_col = asd2[1:len(asd2):2]
    first_col = [item.replace('\t','') for item in first_col]
    temp_table = pd.DataFrame({header[0]:first_col,header[1]:second_col})
    output.append("Cronograma")
    output.append(temp_table)
  else:
    output.append("Cronograma")
    output.append([])
  ##########
  if any("Documentos del Proyecto" in item for item in asd) | any("Otros Documentos" in item for item in asd):
    table2 = wd.find_elements(By.XPATH,"//table[@class='table table-default table-download']//td")
    asd1 = []
    for lnk in table2:
      asd1.append(lnk.get_attribute("innerHTML"))
    first_col = asd1[0:len(table2):3]
    second_col = asd1[1:len(table2):3]
    third_col = asd1[2:len(table2):3]
    link2 = []
    for idx in range(0,len(third_col)):
      link2.append(re.search('"' + '(.+?)' + '"', third_col[idx]).group(1))
    header2 = ["Nombre","Descripción","Link de Descarga"]
    temp_table2 = pd.DataFrame({header2[0]:first_col,header2[1]:second_col,header2[2]:link2})
    output.append("Docs")
    output.append(temp_table2)
  else:
    output.append("Docs")
    output.append([])
```

```python
###OUTPUTS

cronograma = output[2:len(output):5]
docs = output[4:len(output):5]
#proyecto
```

# Exportando datos a txt.

```python
### Exportando cronograma a txt

cronograma2 = []
for idx in range(0,len(cronograma)):
  if isinstance(cronograma[idx], pd.DataFrame):
    cronograma2.append(pd.DataFrame.to_json(cronograma[idx]))
  else:
    cronograma2.append("")
with open('cronograma.txt','w', encoding='utf-8') as f:
    f.write( ';;'.join( cronograma2 ) )
```

```python
### Exportando docs a txt

docs2 = []
for idx in range(0,len(docs)):
  if isinstance(docs[idx], pd.DataFrame):
    docs2.append(pd.DataFrame.to_json(docs[idx]))
  else:
    docs2.append("")
with open('docs.txt','w', encoding='utf-8') as f:
    f.write( ';;'.join( docs2 ) )
```

```python
### Exportando proyectos a txt

with open('proyectos.txt','w', encoding='utf-8') as f:
    f.write( ';;'.join( proyecto ) )
```
