## Filtro de base de datos

1. Quedarse con los hogares Urbanos, asimismo considerar los hogares que completaron la encuesta.

```python
df_urban_main = enapres2020_1.loc[ enapres2020_1.AREA == 'URBANO', : ]
df_urban = df_urban_main.loc[ df_urban_main.RESFIN == 'Completa', : ]
```

2. Verificar si existe duplicados de hogares

```python
df_urban[ df_urban.loc[:, ['CCDD' ,'CCPP' , 'CCDI' ,'CONGLOMERADO' , 'NSELV', 'VIVIENDA', 'HOGAR']].duplicated(keep = False) ]
```

3. Si existiese, filtrar la base de datos para no considerar los duplicados.

```python
df_urban_no_dpl = df_urban[ ~ df_urban.loc[:, ['CCDD' ,'CCPP' , 'CCDI' ,'CONGLOMERADO' , 'NSELV', 'VIVIENDA', 'HOGAR']].duplicated() ].copy()
df_urban_no_dpl
```

## Crear Dummies

1. Crear una Dummy si la zona donde se ubica el hogar instaló cámaras de seguridad. (variable P122A_1). Tenga cuidado con los missing values. Tenga cuidado con los missing values si los hubiera.

```python
enapres2020_1['P122A_1'].unique()  # El vecindaria instaló camaras de seguridad frente a la delincuencia
```

```python
df_urban_no_dpl['Dummy_camara_seguridad'] = np.where(df_urban_no_dpl['P122A_1'] == "Si", 1, np.where(df_urban_no_dpl['P122A_1'].notnull(), 0, np.nan))
```

2. Crear una Dummy si el hogar cuenta con internet (variable P184A_14). Tenga cuidado con los missing values si los hubiera

```python
# Conexión a Internet
enapres2020_1['P184A_14'].unique()
```

```python
df_urban_no_dpl['Dummy_internet'] = np.where(df_urban_no_dpl['P184A_14'] == "¿Conexión a Internet?", 1, np.where(df_urban_no_dpl['P184A_14'].notnull(), 0, np.nan))
```

3. Crear una Dummy si el hogar cuenta con titulo de propiedad (variable P107A)

```python
# Titulos de propiedad 
enapres2020_1['P107A'].unique()
```

```python
df_urban_no_dpl['Dummy_titulo_propiedad'] = np.where(df_urban_no_dpl['P107A'] == "Título de Propiedad", 1, np.where(df_urban_no_dpl['P107A'].notnull(), 0, np.nan))
```

## Group by

1. Obtener el total de hogares por distrito, el total de hogares que cuentan con internet por distrito, finalmente el porcentaje de hogares que cuentan con internet.

```python
# Total de hogares por distrito
base_a= pd.DataFrame(df_urban_no_dpl.groupby( ['CCDD' ,'CCPP', 'CCDI'] ).count().reset_index())
total_hogares = base_a[['CCDD' ,'CCPP','CCDI', 'TOT_HOGAR']]
total_hogares
```

```python
#Total de hogares con internet por distrito
h_i_d = df_urban_no_dpl.groupby( [ 'CCDD' ,'CCPP' , 'CCDI' ] , as_index = False ).agg(distrito_hogar = ('TOT_HOGAR', np.size), internet = ("Dummy_internet", np.sum))
h_i_d
```

```python
# porcentaje de hogares con internet por distrito
h_i_d["hogares_internet_%"] =(h_i_d["distrito_hogar"] / h_i_d["internet"] )*100
h_i_d
```

2. Obtener una base de datos a nivel departamentos con la siguiente información: total de hogares, total de hogares cuyo vecindario instaló cámaras de seguridad y el porcentaje hogares cuyos vecindarios cuentan con cámaras de seguridad. Base_1

```python
Base_1 = df_urban_no_dpl.groupby( ["CCDD"], as_index = False ).agg( departamento_total = ('TOT_HOGAR', np.size), camara_seguridad = ("Dummy_camara_seguridad", np.sum) )
Base_1
```

```python
#Agregar el porcentaje de los hogares con camara por departamento
Base_1["hogar_camara_%"] = (Base_1["departamento_total"] / Base_1["camara_seguridad"])*100
Base_1
```

3. Crear una base de datos a nivel departamentos con la siguiente información: total de hogares que cuentan con títulos de propiedad y porcentaje de hogares que cuentan con títulos de propiedad. Base_2

```python
Base_2 = df_urban_no_dpl.groupby( ["CCDD"], as_index = False).agg(departamento_total = ('TOT_HOGAR', np.size), titulo_total = ("Dummy_titulo_propiedad", np.sum) )
Base_2
```

```python
#Agregar el porcentaje de hogares con titulo de propiedad
Base_2["titulos_propiedad_%"] = (Base_2["titulo_total"] / Base_2["departamento_total"])*100
Base_2
```

## Merge

1. Hacer un merge entre las bases de datos Base_1 y Base_2

```python
Clean_data =Base_1.merge(Base_2, how = 'left', on=["CCDD",'departamento_total'], indicator = False)
Clean_data
```

2. Finalmente guardar la base de datos en el formato SPSS. Llamar a la base de datos (Clean_data)

```python
import pyreadstat

pyreadstat.write_sav(Clean_data,  r"CLEAN_DATA-GRUPO_5-PYTHON.sav")
```
