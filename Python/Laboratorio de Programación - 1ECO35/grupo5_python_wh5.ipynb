```python
import pandas as pd
import numpy as np
import chardet # to get string character format 
import re
```

```python
#Data ENAHO PANEL 2015-2019
data1 = pd.read_stata("../../data/Juntos_program/data.dta")
#Data Programa Juntos
data2 = pd.read_stata("../../data/Juntos_program/unidos.dta")
```

```python
data1
```

### Replicación Clean DHS

```python
data1.columns
```

```python
filter_list = []
filter_list1 = []

#Drop special character '_'
for i in range(0,len(list(data1.columns))):

        w =  re.sub('[\_]','', list(data1.columns)[i])
        filter_list.append(w)
        
        
#Drop last two string
for i in range(0,len(list(filter_list))):

        w =  re.sub(".{2}$",'', list(filter_list)[i])
        filter_list1.append(w)
```

```python
# drop duplicates in a list

new_list = list(dict.fromkeys(filter_list1))
new_list
```

#### Data ENAHO Panel

```python
reshape_panel = pd.wide_to_long(data1, stubnames = new_list, i = ['numper'] , j = 'year' , sep = '_').reset_index()
```

```python
reshape_panel
```

```python
new_list[0] = data1.columns[0]
new_list[1] = data1.columns[1]
new_list.append("year")
data_panel = reshape_panel[new_list]
data_panel = data_panel.rename(columns={'numpanh15': 'numpanh'})
data_panel                               
```

#### MERGE

```python
data2
```

```python
#### Combinamos las bases de PANEL ENAHO y programa Juntos
base_total = pd.merge(data_panel,data2,on='ubigeo')
```

### Porcentaje de distritos por departamentos que se beneficiaron del Programa Juntos en cada año

```python
base_a = base_total.iloc[:,8:13]
base_a
```

```python
base_a = base_a[~base_a.duplicated()]
base_a
```

```python
base_b= pd.DataFrame(base_a.groupby( ['year' ,'region'] ).count().reset_index())

total_distrito = base_b[['year' ,'region','unidos']]

base_c = base_a.groupby( ['year' ,'region'],as_index= False ).agg(sum_juntos = ( "unidos", np.sum) )

base_d = pd.merge(total_distrito,base_c,on=("year","region"))
base_d.rename(columns = {'unidos':'num_distrito'}, inplace = True)
base_d["porc_juntos_depart"]= (base_d["sum_juntos"]  / base_d["num_distrito"]) * 100
base_d
```
